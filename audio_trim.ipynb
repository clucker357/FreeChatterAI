```python
import soundfile as sf
import numpy as np
from matplotlib import pyplot as plt
import librosa
import IPython
src_file1 = "sample_audio_data/BASIC5000_0010.wav"
src_file2 = "sample_audio_data/BASIC5000_0603.wav"
src_file3 = "sample_audio_data/BASIC5000_0701.wav"
src_file3 = "sample_audio_data/BASIC5000_0751.wav"
src_file4 = "sample_audio_data/BASIC5000_0809.wav"
src_file5 = "sample_audio_data/BASIC5000_0907.wav"
src_file6 = "sample_audio_data/BASIC5000_0908.wav"
src_file7 = "sample_audio_data/BASIC5000_0953.wav"
src_file8 = "sample_audio_data/BASIC5000_1524.wav"
src_file9 = "sample_audio_data/BASIC5000_1800.wav"
src_file10 = "sample_audio_data/BASIC5000_1956.wav"
src_file = src_file1
```

```python
data, rate = sf.read(src_file)
t = np.arange(0, len(data))/rate
```

```python
data.shape
```

```python
wav, sr = librosa.load(src_file, sr=None)
display(IPython.display.Audio(wav, rate=sr))
```

```python
wav.shape
```

```python
wave, fs = librosa.load(src_file)
rms = librosa.feature.rms(y=wave)
times = librosa.times_like(rms, sr=fs)
plt.plot(t, data)
plt.plot(times, rms[0]*2**(1/2))
plt.show()
```

```python
y, sr = librosa.load(src_file)
yt, index = librosa.effects.trim(y, top_db=20)
print(librosa.get_duration(y=y), librosa.get_duration(y=yt))
```

```python
t = np.arange(0, len(yt))/rate
plt.plot(t, yt)
plt.show()
```

```python
display(IPython.display.Audio(yt, rate=sr))
```

```python
sf.write('trimmed_sample_data/trimmed.wav', yt, sr)
```
